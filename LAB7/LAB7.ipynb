La matriz de confusión es una herramienta visual que proporciona información sobre qué tan bien se están clasificando los datos según sus etiquetas correspondientes. Al observar una matriz de confusión, los valores en la diagonal representan las clasificaciones correctas. Si nos enfocamos en cualquier punto de la diagonal, los valores verticales son aquellos que fueron clasificados como esa clase, pero en realidad pertenecían a otra clase, mientras que los valores horizontales son las clases que fueron clasificadas incorrectamente y que en realidad correspondían a la clase observada.  

Si la matriz de confusión tiene una alta densidad de datos en la diagonal, significa que el modelo está clasificando correctamente. Por el contrario, si observamos una matriz de confusión con una dispersión alta de datos, significa que el modelo está fallando en su tarea de clasificar.

**Respuesta P4**
Para responder esta pregunta se ocuparan las siguientes abreviaciones:
1. TP: True positive, corresponde a cuando el valor fue clasificado correctamente
2. TN: True Negative, corresponde a cuando el valor no fue clasificado en la clase observada de forma correcta, es decir, el dato corresponde a otra clase
3. FP: False positive, corresponde a cuando el valor fue clasificado como positivo, pero no lo era.
4. FN: False positive, corresponde a cuando el valor fue clasificado como negativo o perteneciente a otra clase, pero correspondia a un positvo o a la clase de observación
5. P: Precision
6. R: Recall


---

**Métricas de evaluación**


*   Accuracy: (TP + TN )/ (TP + TN + FP + FN), la métrica accuracy indica que tan exacto fue el modelo al momento de clasificar los valores ya que divide el numero de predicciones correctas en el número total de predicciones
*   Precision: (TP)/(TP + TF), la métrica precision indica la cantidad de positivos que son True positive en relacion a todos los valores correctamente predichos, esta métrica hace referencia al error de tipo 1
*   Recall (TP)/(TP+ FN), la métrica recall indica la cantidad de positivos que son True Positive en relacion a los valores correctamente catalogados como positivos y aquellos que se clasificaron mal pero eran positivos, esta metrica hace referenia al error de tipo 2
*   F1 score: 1/(α(1/P) + (1-α)(1/R)), donde α generalmente corresponde a 1/2, sin embargo la expresion más usada es 2PR/(P+R)



**Respuestas P5**
Explique qué métrica recomendaría para los siguientes contextos de clasificación. [1 punto cada uno]

*   Mantenimiento predictivo de fallas de maquinaria pesada en la industria minera: En este contexto, es crucial identificar todas las posibles fallas para prevenir interrupciones costosas y peligrosas en la operación minera. No detectar una falla (FN) puede tener consecuencias graves. Por lo tanto, se prefiere la métrica Recall, ya que asegura que la mayoría de las fallas verdaderas sean identificadas, minimizando los falsos negativos.
*   Detección de enfermedades altamente contagiosas: Similar al mantenimiento predictivo, en la detección de enfermedades altamente contagiosas, es vital identificar la mayor cantidad posible de casos verdaderos para evitar la propagación de la enfermedad. Aquí, los falsos negativos (no detectar una persona infectada) pueden tener consecuencias de salud pública muy serias. Por lo tanto, se prioriza el Recall.
* Aprobación de créditos de alto riesgo: En este caso, es importante asegurar que aquellos que son aprobados para un crédito realmente sean buenos candidatos para minimizar el riesgo financiero. La métrica Precision es adecuada porque se enfoca en reducir los falsos positivos, asegurando que la mayoría de los créditos aprobados sean realmente de bajo riesgo.
*  Detección de crímenes: Este contexto puede necesitar un balance entre Precision y Recall, ya que tanto los falsos positivos (acusaciones erróneas) como los falsos negativos (criminales no detectados) son problemáticos. El F1 Score, que es la media armónica de Precision y Recall, proporciona un equilibrio entre ambas métricas, asegurando una adecuada detección de crímenes sin comprometer significativamente ninguna de las dos métricas.

**Respuesta P6**

La calibración de modelos es el proceso de ajustar un modelo predictivo para que sus probabilidades de salida reflejen mejor las probabilidades reales de los eventos que predice. Es decir, se trata de hacer que las predicciones del modelo sean más precisas en términos de probabilidad.

Por ejemplo, si un modelo de clasificación predice que hay un 70% de probabilidad de que un estudiante apruebe un examen, después de la calibración, esta probabilidad debería coincidir lo más posible con la proporción real de estudiantes que aprueban cuando el modelo predice un 70%.

La calibración es importante porque muchas aplicaciones no solo requieren saber la clase más probable, sino también la confianza en esta predicción. Esto es especialmente relevante en áreas como medicina, finanzas, y mantenimiento predictivo, donde entender la probabilidad exacta de un evento puede influir en decisiones críticas.

Se usa para mejorar la toma de decisiones basadas en los modelos predictivos, asegurando que las probabilidades que los modelos proporcionan sean confiables y útiles. Herramientas comunes para la calibración incluyen técnicas como la Platt Scaling y la calibración isotónica. En resumen, la calibración de modelos ayuda a que las predicciones sean más informativas y útiles en la práctica.

````python
# ----------------------------------------------------------------------------------------------------------------------------
# 1. Separación en subconjuntos de entrenamiento y testeo.
df.reset_index(drop=True)
# El parámetro stratify permite que las proporciones de las clases en el conjunto de datos original se mantenga en los subconjuntos de entrenamiento y testeo.
X_train, X_test, y_train, y_test = train_test_split(df,y, test_size=0.3, random_state=0, stratify=y)

# ----------------------------------------------------------------------------------------------------------------------------
# 2. Definicion de pipelines
pipeline_xgboost = Pipeline([('column_transformer', column_transformer), ('clf', GradientBoostingClassifier())])
pipeline_lightbm = Pipeline([('column_transformer', column_transformer),('clf', ltb.LGBMClassifier())])

# ----------------------------------------------------------------------------------------------------------------------------
# 3. Entrenamiento de pipelines
# Pipeline XGBOOST
xgboost_fit = pipeline_xgboost.fit(X_train, y_train)
xgboost_predict = xgboost_fit.predict(X_test)

# Pipeline XGBOOST
lgbm_fit = pipeline_lightbm.fit(X_train, y_train)
lgbm_predict = lgbm_fit.predict(X_test)
````
