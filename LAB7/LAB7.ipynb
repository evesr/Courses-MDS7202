La matriz de confusi√≥n es una herramienta visual que proporciona informaci√≥n sobre qu√© tan bien se est√°n clasificando los datos seg√∫n sus etiquetas correspondientes. Al observar una matriz de confusi√≥n, los valores en la diagonal representan las clasificaciones correctas. Si nos enfocamos en cualquier punto de la diagonal, los valores verticales son aquellos que fueron clasificados como esa clase, pero en realidad pertenec√≠an a otra clase, mientras que los valores horizontales son las clases que fueron clasificadas incorrectamente y que en realidad correspond√≠an a la clase observada.  

Si la matriz de confusi√≥n tiene una alta densidad de datos en la diagonal, significa que el modelo est√° clasificando correctamente. Por el contrario, si observamos una matriz de confusi√≥n con una dispersi√≥n alta de datos, significa que el modelo est√° fallando en su tarea de clasificar.

**Respuesta P4**
Para responder esta pregunta se ocuparan las siguientes abreviaciones:
1. TP: True positive, corresponde a cuando el valor fue clasificado correctamente
2. TN: True Negative, corresponde a cuando el valor no fue clasificado en la clase observada de forma correcta, es decir, el dato corresponde a otra clase
3. FP: False positive, corresponde a cuando el valor fue clasificado como positivo, pero no lo era.
4. FN: False positive, corresponde a cuando el valor fue clasificado como negativo o perteneciente a otra clase, pero correspondia a un positvo o a la clase de observaci√≥n
5. P: Precision
6. R: Recall


---

**M√©tricas de evaluaci√≥n**


*   Accuracy: (TP + TN )/ (TP + TN + FP + FN), la m√©trica accuracy indica que tan exacto fue el modelo al momento de clasificar los valores ya que divide el numero de predicciones correctas en el n√∫mero total de predicciones
*   Precision: (TP)/(TP + TF), la m√©trica precision indica la cantidad de positivos que son True positive en relacion a todos los valores correctamente predichos, esta m√©trica hace referencia al error de tipo 1
*   Recall (TP)/(TP+ FN), la m√©trica recall indica la cantidad de positivos que son True Positive en relacion a los valores correctamente catalogados como positivos y aquellos que se clasificaron mal pero eran positivos, esta metrica hace referenia al error de tipo 2
*   F1 score: 1/(Œ±(1/P) + (1-Œ±)(1/R)), donde Œ± generalmente corresponde a 1/2, sin embargo la expresion m√°s usada es 2PR/(P+R)

**Respuesta P6**

La calibraci√≥n de modelos es el proceso de ajustar un modelo predictivo para que sus probabilidades de salida reflejen mejor las probabilidades reales de los eventos que predice. Es decir, se trata de hacer que las predicciones del modelo sean m√°s precisas en t√©rminos de probabilidad.

Por ejemplo, si un modelo de clasificaci√≥n predice que hay un 70% de probabilidad de que un estudiante apruebe un examen, despu√©s de la calibraci√≥n, esta probabilidad deber√≠a coincidir lo m√°s posible con la proporci√≥n real de estudiantes que aprueban cuando el modelo predice un 70%.

La calibraci√≥n es importante porque muchas aplicaciones no solo requieren saber la clase m√°s probable, sino tambi√©n la confianza en esta predicci√≥n. Esto es especialmente relevante en √°reas como medicina, finanzas, y mantenimiento predictivo, donde entender la probabilidad exacta de un evento puede influir en decisiones cr√≠ticas.

Se usa para mejorar la toma de decisiones basadas en los modelos predictivos, asegurando que las probabilidades que los modelos proporcionan sean confiables y √∫tiles. Herramientas comunes para la calibraci√≥n incluyen t√©cnicas como la Platt Scaling y la calibraci√≥n isot√≥nica. En resumen, la calibraci√≥n de modelos ayuda a que las predicciones sean m√°s informativas y √∫tiles en la pr√°ctica.

**Respuesta:**

````python
# ----------------------------------------------------------------------------------------------------------------------------
# 1. Separaci√≥n en subconjuntos de entrenamiento y testeo.
df.reset_index(drop=True)
# El par√°metro stratify permite que las proporciones de las clases en el conjunto de datos original se mantenga en los subconjuntos de entrenamiento y testeo.
X_train, X_test, y_train, y_test = train_test_split(df,y, test_size=0.3, random_state=0, stratify=y)

# ----------------------------------------------------------------------------------------------------------------------------
# 2. Definicion de pipelines
pipeline_xgboost = Pipeline([('column_transformer', column_transformer), ('clf', GradientBoostingClassifier())])
pipeline_lightbm = Pipeline([('column_transformer', column_transformer),('clf', ltb.LGBMClassifier())])

# ----------------------------------------------------------------------------------------------------------------------------
# 3. Entrenamiento de pipelines
# Pipeline XGBOOST
xgboost_fit = pipeline_xgboost.fit(X_train, y_train)
xgboost_predict = xgboost_fit.predict(X_test)

# Pipeline XGBOOST
lgbm_fit = pipeline_lightbm.fit(X_train, y_train)
lgbm_predict = lgbm_fit.predict(X_test)
````
